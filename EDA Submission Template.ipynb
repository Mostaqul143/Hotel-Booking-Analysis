# **Project Summary -**

The dataset provides information about hotels and their characteristics, including location, amenities, and prices, as well as customer ratings and reviews. Exploratory data analysis of the dataset reveals several key insights that can be used to improve hotel performance and customer satisfaction.

One of the key insights from the dataset is the strong relationship between hotel ratings and the number of reviews. Hotels with more reviews tend to have higher ratings, which suggests that customer feedback and engagement are important factors in driving customer satisfaction. Hotels may be able to improve their ratings by actively soliciting and responding to customer reviews, as well as by improving the overall customer experience.

Another important factor in customer satisfaction is the price of the hotel. The dataset reveals that customers are highly sensitive to price, with demand dropping significantly as prices increase. However, the dataset also suggests that there may be opportunities for hotels to optimize their pricing strategies by offering discounts and promotions during off-peak periods or for specific types of customers.

The dataset also reveals several interesting trends in customer behavior and preferences. For example, customers tend to prefer hotels that are located near tourist attractions and that offer amenities such as free Wi-Fi and breakfast. Customers also tend to book hotels closer to the date of their stay, suggesting that hotels may be able to capture additional revenue by offering last-minute deals or promotions.

In terms of market differentiation, the dataset suggests that there may be opportunities for hotels to differentiate themselves by offering unique amenities or targeting specific market segments. For example, hotels that cater to families may benefit from offering child-friendly amenities such as pools and playgrounds, while hotels that cater to business travelers may benefit from offering meeting rooms and other business-focused amenities.

Overall, the dataset provides valuable insights into the factors that drive customer satisfaction and hotel performance. By leveraging these insights, hotels can optimize their pricing strategies, improve their service offerings and amenities, and better target their marketing and customer engagement efforts to improve customer satisfaction and drive revenue growth.

## ***2. Understanding Your Variables***

````python
# Dataset Columns

df.columns
````

### Variables Description 

*    **hotel:** Hotel (H1 = Resort Hotel or H2 = City Hotel)

*    **is_canceled:** Value indicating if the booking was canceled (1) or not (0)

*    **lead_time:** Number of days between the date of booking and the arrival date

*    **arrival_date_year:** Year of arrival date.

*    **arrival_date_month:** Month of arrival date

*    **arrival_date_week_number:** Week number of year for arrival date

*    **arrival_date_day_of_month:** Day of month for arrival date

*    **stays_in_weekend_nights:** Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel

*    **stays_in_week_nights:** Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel

*    **adults:** Number of adults

*    **children:** Number of children

*    **babies:** Number of babies

*    **meal:** Type of meal booked

*    **country:** Country of origin

*    **market_segment:** Market segment designation

*    **distribution_channel:** Booking distribution channel

*    **is_repeated_guest:** Value indicating if the booking name was from a repeated guest (1) or not (0)

*    **previous_cancellations:** Number of previous bookings that were canceled by the customer prior to the current booking

*    **previous_bookings_not_canceled:** Number of previous bookings not canceled by the customer prior to the current booking

*    **reserved_room_type:** Code of room type reserved

*    **assigned_room_type:** Code for the type of room assigned to the booking

*    **booking_changes:** Number of changes made to the booking from the moment the booking was entered into the PMS until the moment of check-in or  cancellation

*    **deposit_type:** Type of deposit made by the customer

*    **agent:** ID of the travel agency that made the booking

*    **company:** ID of the company/entity that made the booking or responsible for paying the booking

*    **days_in_waiting_list:** Number of days the booking was in the waiting list before it was confirmed to the customer

*    **customer_type:** Type of booking, assuming one of four categories: Contract, Group, Transient, and Transient-Party

*    **adr:** Average daily rate (revenue per room)

*    **required_car_parking_spaces:** Number of car parking spaces required by the customer

*    **total_of_special_requests:** Number of special requests made by the customer (e.g. twin bed or high floor)
